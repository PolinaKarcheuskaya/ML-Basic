## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
from datetime import datetime

def date_check(date: str) -> bool:
    try:
        datetime.strptime(date, '%d.%m.%Y')
        return True
    except ValueError:
        return False

print(date_check('29.02.2000'))
print(date_check('29.02.2001'))
print(date_check('31.04.1962'))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
import math

def simple_number_check(number: int) -> bool:
    if number < 2:
        return False
    if number == 2:
        return True
    if number % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False
    return True

print(simple_number_check(17))
print(simple_number_check(20))
print(simple_number_check(23))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
users = {}

def add_user(id: str, name: str, surname: str, age: int) -> str:
    if not name.isalpha() or not surname.isalpha():
        return 'Имя и фамилия должны состоять из букв'
    name = name.capitalize()
    surname = surname.capitalize()
    if age < 18 or age > 60:
        return 'Возраст должен быть числом от 18 до 60'
    id = id.zfill(8)
    if id in users:
        return 'Id существует'
    else:
        users[id] = (name, surname, age)
        return 'Пользователь добавлен'

print('Если хотите закончить ввод, введите пустую строку')
while True:
    id = input('Введите ID: ')
    if id == '':
        break
    name = input('Введите имя: ')
    surname = input('Введите фамилию: ')
    age = int(input('Введите возраст: '))
    add_user(id, name, surname, age)

def print_users(users: dict) -> str:
    print('{:<10} {:<15} {:<15} {:<5}'.format('ID', 'Имя', 'Фамилия', 'Возраст'))
    for id, user in users.items():
        print('{:<10} {:<15} {:<15} {:<5}'.format(id, user[0], user[1], user[2]))

print_users(users)

```
