# Линейная алгебра, математический анализ

Исследовательская задача: низкоранговое приближение матрицы

1. Сгенерировать случайную квадратную матрицу A размера n×n, где n ≥ 100.
2. Выполнить SVD-разложение матрицы A с помощью numpy.linalg.svd, получить U, S и V^T.
3. Для рангов r от 2 до n построить приближение Ã = U[:, :r] S[:r, :r] V^T[:r, :].
4. Для каждого r вычислить ошибку приближения E(r) = ||A − Ã||_F.
5. Построить график зависимости ошибки E(r) от ранга r.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(88)

n = 100
A = np.random.randn(n, n)

U, S, Vt = np.linalg.svd(A)

ranks = np.arange(2, n + 1)
errors = []

for r in ranks:
    A_tilde = U[:, :r] @ np.diag(S[:r]) @ Vt[:r, :]
    errors.append(np.linalg.norm(A - A_tilde, ord='fro'))

errors = np.array(errors)

plt.figure(figsize=(10, 6))
plt.plot(ranks, errors, marker='o', markersize=1)
plt.xlabel('Ранг r')
plt.ylabel('Ошибка E(r)')
plt.title('Зависимость ошибки низкорангового приближения от ранга')
plt.grid(True)
plt.show()
```
